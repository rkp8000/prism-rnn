# Legendre expansion of arcsin

Starting from the following identity (https://www.cfm.brown.edu/people/dobrush/am34/Mathematica/ch5/legendre.html):

$$\frac{1}{\sqrt{1 - x^2}} = \frac{\pi}{2}\sum_{n \geq 0} (4n+1) \left[\frac{(2n-1)!!}{n!2^n}\right]^2 P_{2n}(x)$$

$$
= \frac{\pi}{2}\left[
P_0(x) + \sum_{n \geq 1} (4n+1) \left[\frac{(2n-1)!!}{n!2^n}\right]^2 P_{2n}(x)
\right]
$$

$$
\arcsin x = \int \frac{1}{\sqrt{1-x^2}}$$

Using the identity

$$\int P_l(x) dx = \frac{P_{l+1}(x) - P_{l-1}(x)}{2l+1} \quad \quad \forall \; l \geq 1,$$

we have

$$
\arcsin x = \frac{\pi}{2}\left[
    P_1(x) + \sum_{n \geq 1} \left[\frac{(2n-1)!!}{n!2^n}\right]^2
    \left[P_{2n+1}(x) - P_{2n-1}(x)\right]
\right]
$$

$$
= \frac{\pi}{2}\left[
    P_1(x) + \sum_{n \geq 1} \left[\frac{(2n-1)!!}{n!2^n}\right]^2
    P_{2n+1}(x) - \sum_{n \geq 1} \left[\frac{(2n-1)!!}{n!2^n}\right]^2P_{2n-1}(x)
\right]
$$

Letting $n' = n-1$, we have

$$
\arcsin x = \frac{\pi}{2}\left[
    P_1(x) + \sum_{n \geq 1} \left[\frac{(2n-1)!!}{n!2^n}\right]^2 P_{2n+1}(x)
    - \sum_{n' \geq 0} \left[\frac{(2n'+1)!!}{(n'+1)!2^{n'+1}}\right]^2P_{2n'+1}(x)
\right]
$$

$$
= \frac{\pi}{2}\left[\frac{1}{2}P_1(x)
+ \sum_{n \geq 1}\left(
    \left[\frac{(2n-1)!!}{n!2^n}\right]^2 - \left[\frac{(2n+1)!!}{(n+1)!2^{n+1}}\right]^2
\right)P_{2n+1}(x)
\right]
$$

$$
\frac{(2n+1)!!}{(n+1)!2^{n+1}}
=
\frac{2n+1}{2n+2}\frac{(2n-1)!!}{n!2^n} 
< 
\frac{(2n-1)!!}{n!2^n} 
$$

Hence
$$
\left[\frac{(2n+1)!!}{(n+1)!2^{n+1}}\right]^2
< 
\left[\frac{(2n-1)!!}{n!2^n} \right]^2
$$
so the coefficients in the sum are positive.
