```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.special import sph_harm
from sklearn.linear_model import Ridge
```

```python
n = 1000

rth = np.linspace(0, np.pi, n)
k = 1 - rth/np.pi

n_basis = 50
zonals = np.array([
    np.real(sph_harm(0, l, np.zeros(n), rth)) for l in range(n_basis)
])
```

```python
fig, axs = plt.subplots(1, 3, figsize=(9, 2), tight_layout=True)
axs[0].plot(rth, k)
[axs[1].plot(rth, zonal) for zonal in zonals[0:10:2]]
[axs[2].plot(rth, zonal) for zonal in zonals[1:5:2]]

fig, axs = plt.subplots(1, 3, figsize=(9, 2), tight_layout=True)
axs[0].plot(rth, k)
[axs[1].plot(rth, zonal*np.sin(rth)) for zonal in zonals[0:10:2]]
[axs[2].plot(rth, zonal*np.sin(rth)) for zonal in zonals[1:11:2]]

fig, axs = plt.subplots(1, 3, figsize=(13, 4), tight_layout=True)
axs[0].plot(rth, k)
[axs[1].plot(rth, (k-.5)*zonal*np.sin(rth)) for zonal in zonals[0:10:2]]
[axs[2].plot(rth, (k-.5)*zonal*np.sin(rth)) for zonal in zonals[1:11:2]]
axs[2].grid()

for l, zonal in enumerate(zonals):
    print('l=', l, 'w_l=', np.sum((k-.5)*zonal*np.sin(rth)))
```

```python
rgr = Ridge(alpha=0, fit_intercept=False).fit((zonals).T, k)
w = rgr.coef_
```

```python
fig, axs = plt.subplots(1, 2, figsize=(10, 6), tight_layout=True)
axs[0].plot(w)

axs[0].grid()
axs[1].plot(rth, zonals.T@w)
```
