Numerically check formula that

$$\arcsin(x) = \frac{2^n(n!)^2}{(2n+1)(2n!)} P_{2n+1}(x)$$

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.special import legendre, factorial, factorial2

fac = factorial
fac2 = lambda n: 1 if n == -1 else factorial2(n)

rx = np.linspace(-1., 1, 1000)
dx = np.mean(np.diff(rx))

lps = np.array([legendre(l)(rx) for l in range(100)])
```

```python
arcsin = np.arcsin(rx)

fig, ax = plt.subplots(1, 1, figsize=(8, 8), tight_layout=True)
ax.plot(rx, arcsin, lw=3)

approx = np.zeros(len(rx))

for n in range(50):
    # from brown university
    num = fac2(2*n-1)
    denom = fac(n)*(2**n)

    approx += (np.pi/2) * ((num/denom)**2) * lps[2*n+1]
    
    ax.plot(rx, approx, lw=1)

ax.grid()
```
