Rand-tanh kernel

$$k(x, y) = \boldsymbol{\psi}(\mathbf{x})^T\boldsymbol{\psi}(\mathbf{y})$$

$$\boldsymbol{\psi}(\mathbf{x}) = \tanh J\mathbf{x}$$

$$J_{ij} \sim \mathcal{N}(\cdot| 0, g^2)$$

What is the average kernel?

$$\bar{k}(\mathbf{x}, \mathbf{y}) = \mathbb{E}_J [k(\mathbf{x}, \mathbf{y})] = ?$$

$$J\mathbf{x} = [J_{11}x_1 + \dots + J_{1D}x_D, \dots, J_{N1}x_1 + \dots J_{ND}x_D]^T$$

$$\boldsymbol{\psi}(\mathbf{x}) = \textbf{tanh}(J\mathbf{x})$$

$$
=
\begin{bmatrix}
\tanh(J_{11}x_1 + \dots + J_{1D}x_D) \\
\vdots \\
\tanh(J_{N1}x_1 + \dots + J_{ND}x_D) \\
\end{bmatrix}
$$

$$k(\mathbf{x}, \mathbf{y}) = \textbf{tanh}(Jx)^T \textbf{tanh}(Jy) = $$

$$ 
\begin{matrix}
\tanh(J_{11}x_1 + \dots + J_{1D}x_D)\tanh(J_{11}y_1 + \dots + J_{1D}y_D) \\
+ \dots + \\
\tanh(J_{N1}x_1 + \dots + J_{ND}x_D)\tanh(J_{N1}y_1 + \dots + J_{ND}y_D) \\
\end{matrix}
$$

$$\mathbf{J}_{i, :}\mathbf{x} \sim \mathcal{N}(\cdot| 0, \sigma_x^2)$$

$$\mathbf{J}_{i, :}\mathbf{y} \sim \mathcal{N}(\cdot| 0, \sigma_y^2)$$

$$\mathbb{E}[k(\mathbf{x}, \mathbf{y})] = $$

$$
\begin{matrix}
\mathbb{E}[\tanh(J_{11}x_1 + \dots + J_{1D}x_D)\tanh(J_{11}y_1 + \dots + J_{1D}y_D)] \\
+ \dots + \\
\mathbb{E}[\tanh(J_{N1}x_1 + \dots + J_{ND}x_D)\tanh(J_{N1}y_1 + \dots + J_{ND}y_D)] \\
\end{matrix}
$$

$$
= N\mathbb{E}[\tanh(J_{11}x_1 + \dots + J_{1D}x_D)\tanh(J_{11}y_1 + \dots + J_{1D}y_D)] \\
$$


Linear limit ($\tanh(x) \approx x$):

$$\mathbb{E} = N\mathbb{E}[(J_{11}x_1 + \dots + J_{1D}x_D)(J_{11}y_1 + \dots + J_{1D}y_D)]$$

$$
=N\left(\mathbb{E}[
J_{11}^2 x_1 y_1 + \dots + J_{1D}^2 x_Dy_D 
]
+ \mathbb{E}[
J_{11} J_{12} x_1 y_2 + \dots
]
\right)
$$ 

$$
=N\mathbb{E}[
J_{11}^2 x_1 y_1 + \dots + J_{1D}^2 x_Dy_D 
]
$$

$$
=Ng^2(x_1y_1 + \dots + x_Dy_D)
